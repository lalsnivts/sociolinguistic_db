```python
import folium
import pandas as pd
```

```python
df = pd.read_excel('coordinates_edited.xlsx')
```

```python
df.head(1)
```

```python
m = folium.Map(location=[63.899, 95.889],
              zoom_start=3)
```

```python
for num, row in df.iterrows():
    folium.Circle(
                location=[row['latitude'], row['longitude']],
                popup=row['name']).add_to(m)
```
