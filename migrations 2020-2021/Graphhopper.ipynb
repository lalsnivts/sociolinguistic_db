Ограничения: https://www.graphhopper.com/pricing/

```python
import requests
import json
import folium
```

```python
api_key = '180fb566-3b1d-41a5-a465-c27ba99fdb96'
```

```python
response = requests.get("https://graphhopper.com/api/1/route?point=59.97288249898143,30.311279296875&point=55.91842985630817,37.716064453125&vehicle=car&calc_points=true&points_encoded=false&key={}".format(
api_key))
response = json.loads(response.text)
```

```python
r = response['paths'][0]['points']['coordinates']
```

```python
r = [[i[1], i[0]] for i in r]
```

```python
m = folium.Map(location=[59.98259700478887,30.4297523923562],
              zoom_start=5)
```

```python
folium.PolyLine(r,
                color='red',
                weight=15,
                opacity=0.8).add_to(m)
```
